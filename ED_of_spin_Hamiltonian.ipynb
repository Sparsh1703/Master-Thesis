```python
from quspin.operators import hamiltonian
from quspin.basis import spin_basis_1d
import numpy as np
```

```python
# define model parameters
L = 12
Jxy = np.sqrt(2.0)
Jzz_0 = 1.0
hz = 1.0/np.sqrt(3.0)
```

```python
basis = spin_basis_1d(L, pauli=False)
basis_1 = spin_basis_1d(L, pauli=False, Nup = L//2)
basis_2 = spin_basis_1d(L, pauli=False, Nup = L//2, pblock = 1)
```

```python
J_zz = [[Jzz_0,i,i+1] for i in range(L-1)]
J_xy = [[Jxy/2.0,i,i+1] for i in range(L-1)]
h_z = [[hz,i] for i in range (L-1)]
```

```python
static = [["+-", J_xy],["-+",J_xy],["zz",J_zz]]
dynamic = []
```

```python
H_XXZ = hamiltonian(static,dynamic, basis = basis, dtype = np.float64)
E = H_XXZ.eigvalsh()
```
